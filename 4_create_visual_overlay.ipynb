```python
## All dimensions must match, all must be 1 channel, all mask files must have 1 value per mask
## MAKE SURE TO ALWAYS LEAVE ONE CONSTANT AS HCR 1, DON'T SWITCH THE IDENTITIES
#Output dataframes (dfs) saved in the HCR round folder in NASQUATCH for now


stack1_image_path = fr"\\bananas\data\ezfish\jonna\CIM132_bk\2P\tile\single_plane_bigwarp_previous\CIM132_001L_HighRes_Stitched_8bit_Composite_RotMinus90_with_black_planes_highres_BigWarp_OUTPUT_c1.tif" #Must be 1 channel!!!
stack1_masks_path = fr"\\bananas\data\ezfish\jonna\CIM132_bk\2P\tile\single_plane_bigwarp_previous\MASKS_LowToHigh_HighToHCR_BigWarp_OUTPUT_80distancegood.tif"

#ALWAYS LEAVE THIS AS HCR ROUND 1 (TARGET ROUND)
stack2_image_path = fr"\\bananas\data\ezfish\jonna\CIM132_bk\HCR\240408_CIM132\HCR1_Masks\CIM132_SLICE001_HCR1_C1_DAPI.tif" #Must be 1 channel!!!
stack2_masks_path = fr"\\bananas\data\ezfish\jonna\CIM132_bk\HCR\240408_CIM132\HCR1_Masks\CIM132_SLICE001_HCR1_cp_masks.tif"

min_overlap = 0.3

output_base_filename = fr"\\bananas\data\ezfish\jonna\CIM132_bk\2P\Visualization_min_overlap_{min_overlap}"

# calculate the matching masks and the overlap
mask1_to_mask2_df = match_masks(stack1_masks_path, stack2_masks_path)

# filter only overlaps above min_overlap
mask1_to_mask2 = mask1_to_mask2_df.query(f'overlap>@min_overlap')[['mask1','mask2']].astype('int').set_index('mask1')['mask2'].to_dict()

# creating visualization files
visualize_match(stack1_image_path, stack1_masks_path, stack2_image_path, stack2_masks_path,
                mask1_to_mask2, output_base_filename)
```
